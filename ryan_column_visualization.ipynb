```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('merge_data_update.csv')
```

```python
data
```

```python
pd.concat([data['SalePrice'], data.loc[:, 'Heating':'FireplaceQu']], axis=1)
```

```python
ryan_cols = data.loc[:, 'Heating':'FireplaceQu'].columns
ryan_cols
```

```python
print('Column Name: ', ryan_cols[0])
# Heating column visualization
#     Floor - Floor Furnace
#     GasA - Gas forced warm air furnace
#     GasW - Gas hot water or steam heat
#     Grav - Gravity furnace
#     OthW - Hot water or steam heat other than gas
#     Wall - Wall furnace
print('Number of Null Values: ',np.sum(data['Heating'].isnull()))
print(data['Heating'].value_counts())
data['Heating'].hist(weights=np.ones(len(data['Heating']))/len(data['Heating']))
plt.show()
sns.boxplot(x='Heating', y='SalePrice', data=data)
```

```python
print('Column Name: ', ryan_cols[1])
# HeatingQC: Heating quality and condition

#        Ex - Excellent
#        Gd - Good
#        TA - Average/Typical
#        Fa - Fair
#        Po - Poor
print('Number of Null Values: ',np.sum(data['HeatingQC'].isnull()))
print(data['HeatingQC'].value_counts())
data['HeatingQC'].hist(weights=np.ones(len(data['HeatingQC']))/len(data['HeatingQC']))
plt.show()
sns.boxplot(x='HeatingQC', y='SalePrice', data=data, order=['Ex', 'Gd', 'TA', 'Fa', 'Po'])
```

```python
print('Column Name: ', ryan_cols[2])
# CentralAir: Central air conditioning
#     N - No
#     Y - Yes
print('Number of Null Values: ',np.sum(data['CentralAir'].isnull()))
print(data['CentralAir'].value_counts())
data['CentralAir'].hist(weights=np.ones(len(data['CentralAir']))/len(data['CentralAir']))
plt.show()
sns.boxplot(x='CentralAir', y='SalePrice', data=data)
```

```python
print('Column Name: ', ryan_cols[3])
# Electrical: Electrical system
#     SBrkr - Standard Circuit Breakers & Romex
#     FuseA - Fuse Box over 60 AMP and all Romex wiring (Average)	
#     FuseF - 60 AMP Fuse Box and mostly Romex wiring (Fair)
#     FuseP - 60 AMP Fuse Box and mostly knob & tube wiring (poor)
#     Mix - Mixed
print('Number of Null Values: ',np.sum(data['Electrical'].isnull()))
print(data['Electrical'].value_counts())
data['Electrical'].hist(weights=np.ones(len(data['Electrical'])-1)/len(data['Electrical'])) # The -1 is for the Null value
plt.show()
sns.boxplot(x='Electrical', y='SalePrice', data=data, order=['SBrkr', 'FuseA', 'FuseF', 'FuseP', 'Mix'])

## Maybe fill in the null based on year built ##
```
