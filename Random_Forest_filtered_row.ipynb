```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
```

```python
# import test and train data sets

data_test = pd.read_csv('RowFiltered_dummied_data_TEST.csv')
data_train = pd.read_csv('RowFiltered_dummied_data_TRAIN.csv')
```

```python
X_train = data_train.iloc[:, 1:]
X_test = data_test.iloc[:,1:]
y_train = data_train.iloc[:,0]
y_test  = data_test.iloc[:,0]
```

```python
y_train.max()
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
rf = RandomForestRegressor(n_jobs=1, random_state=0, verbose=2)

rf.fit(X_train, y_train)

print("The training error is : %.5f" % (1 - rf.score(X_train, y_train)))
print("The test error is: %.5f" % (1 - rf.score(X_test, y_test)))
```

```python
from sklearn.model_selection import cross_val_score
cv_score = cross_val_score(rf, X_train, y_train, cv=5)
cv_score.mean()
```

```python
y_pred = rf.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

```python
rf.get_params()
```

```python
random_param = {'min_samples_split': [2, 5, 10, 15],
                'max_features': ['sqrt', .5, 'auto'],
                'n_estimators': [150, 300, 450],
                
}
```

```python
from sklearn.model_selection import GridSearchCV
rf_grid = GridSearchCV(estimator = rf, param_grid = random_param,
                                cv=5, return_train_score=True)

rf_grid.fit(X_train, y_train)
```

```python
rf_grid.best_score_
```

```python
rf_grid.best_estimator_
```

```python
rf_grid.best_params_
```

```python
rf.set_params(max_features=.5, min_samples_split= 2, n_estimators=450, oob_score=True)
```

```python
rf.fit(X_train, y_train)

print("The new training error is : %.5f" % (1 - rf.score(X_train, y_train)))
print("The new test error is: %.5f" % (1 - rf.score(X_test, y_test)))
```

```python
rf.oob_score_
```

```python
cv_new_score = cross_val_score(rf, X_train, y_train, cv=5)
cv_new_score.mean()
```
